```python
#First Innings Score Predictor for Indian Premier League (IPL)
import pandas as pd

df = pd.read_csv('IPL_Data.csv')

df.head()

len(df['venue'].unique())

#Data Cleaning --
#Removing unwanted Columns --
columns_to_remove = ['mid','venue','batsman','bowler','striker','non-striker']
df = df.drop( labels = columns_to_remove, axis = 1)


df.head()
df.bat_team.unique()

regular_teams = ['Kolkata Knight Riders','Chennai Super Kings','Mumbai Indians',

                 'Rajasthan Royals','Kings XI Punjab','Delhi Daredevils',

                 'Sunrisers Hyderabad','Royal Challengers Bangalore']

df = df[ (df['bat_team'].isin(regular_teams)) & (df['bowl_team'].isin(regular_teams))]

df.head()
df = df[df['overs'] >= 5.0]

print(df.bat_team.unique())

print(len(df.bat_team.unique()))

# To Convert Date --



from datetime import datetime

df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))

df.head()
# Filtering the given data df --



df_new = pd.get_dummies(data=df, columns=['bat_team','bowl_team'])

df_new.head()
df_new = df_new[['date', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',

                 'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',

                 'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',

                 'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',

                 'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',

                 'bowl_team_Royal Challengers Bangalore', 'bowl_team_Sunrisers Hyderabad',

                 'overs', 'runs', 'wickets', 'runs_last_5', 'wickets_last_5', 'total']]
#creating train and test data :-

X_train = df_new.drop(labels = 'total', axis = 1)[df_new['date'].dt.year <= 2016]



X_test = df_new.drop(labels = 'total', axis = 1)[df_new['date'].dt.year >= 2017]

Y_train = df_new[df_new['date'].dt.year <= 2016]['total'].values



Y_test = df_new[df_new['date'].dt.year >= 2017]['total'].values

X_train.drop(labels = 'date', axis=True, inplace=True)

X_test.drop(labels = 'date', axis=True, inplace= True)
#Using LinearRegression to test and predict the result

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, Y_train)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None,
         normalize=False)

prediction = regressor.predict(X_test)

from sklearn import metrics

import numpy as np

print('MAE:', metrics.mean_absolute_error(Y_test, prediction))

print('MSE:', metrics.mean_squared_error(Y_test, prediction))

print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, prediction)))
#Using Lasso to test and predict the result:

from sklearn.linear_model import Lasso

from sklearn.model_selection import GridSearchCV

lasso = Lasso()

parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}

lasso_regressor = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)



lasso_regressor.fit(X_train,Y_train)

print(lasso_regressor.best_params_)

print(lasso_regressor.best_score_)
# Importing Pickle --

import pickle

filename = 'IPL_Model.pkl'

pickle.dump(regressor, open(filename, 'wb'))

prediction = lasso_regressor.predict(X_test)

from sklearn import metrics

import numpy as np

print('MAE:', metrics.mean_absolute_error(Y_test, prediction))

print('MSE:', metrics.mean_squared_error(Y_test, prediction))

print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, prediction)))
```
